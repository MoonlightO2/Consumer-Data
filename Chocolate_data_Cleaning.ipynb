## **Dataset 2 - Chocolate**

```python
data2 = pd.read_csv(file2)
```

```python
# Assuming data2 is your DataFrame
data2 = data2.drop('Unnamed: 0', axis=1)
```

```python
data2.rename(columns={
    'ref': 'Ref',
    'company': 'Company',
    'company_location': 'Location',
    'review_date': 'Review Date',
    'country_of_bean_origin': 'Bean Origin',
    'specific_bean_origin_or_bar_name': 'Specific Bean Origin or Bar Name',
    'cocoa_percent': 'Cocoa Percent',
    'rating': 'Rating',
    'counts_of_ingredients': 'Counts of Ingredients',
    'beans': 'Beans',
    'cocoa_butter': 'Cocoa Butter',
    'vanilla': 'Vanilla',
    'lecithin': 'Lecithin',
    'salt': 'Salt',
    'sugar': 'Sugar',
    'sweetener_without_sugar': 'Sweetener without Sugar',
    'first_taste': 'Taste 1',
    'second_taste': 'Taste 2',
    'third_taste': 'Taste 3',
    'fourth_taste': 'Taste 4'
}, inplace=True)
```

```python
data2 = data2.applymap(lambda x: x.title().replace('_', ' ').strip() if isinstance(x, str) else x)
```

```python
# Define a dictionary to map old values to new values for each column
rename_mapping = {
    'Beans': {'Have Bean': 'Bean'},
    'Cocoa Butter': {'Have Cocoa Butter': 'Yes', 'Have Not Cocoa Butter': 'No'},
    'Vanilla': {'Have Not Vanila': 'No', 'Have Vanila': 'Yes'},
    'Lecithin': {'Have Not Lecithin': 'No', 'Have Lecithin': 'Yes'},
    'Salt': {'Have Not Salt': 'No', 'Have Salt': 'Yes'},
    'Sugar': {'Have Sugar': 'Yes', 'Have Not Sugar': 'No'},
    'Sweetener without Sugar': {'Have Not Sweetener Without Sugar': 'No', 'Have Sweetener Without Sugar': 'Yes'}
}

# Iterate over each column and apply the mapping
for col, mapping in rename_mapping.items():
    data2[col] = data2[col].replace(mapping)
```

```python
cols = ['Beans', 'Cocoa Butter', 'Vanilla', 'Lecithin', 'Salt', 'Sugar', 'Sweetener without Sugar']

for col in cols:
  unique_count = data2[col].nunique()
  print(f"Unique values in '{col}':", unique_count)
  print(f"Unique values in '{col}':", data2[col].unique(), '\n')
```

```python
data2.to_csv(file5)
print('Cleaned file creation successful...!!', file5)
```

```python
print('Shape: ', data2.shape, '\n')
print('Columns: ', data2.columns.tolist(), '\n')
data2.head()
```
